## Предобработка данных

### Обработка пропусков

```python
# посчитаем пропущенные значения и отсортируем их
data.isna().sum().sort_values(ascending = False) 
```

Довольно много пропусков, выявим причины пропусков для каждого столбца по-отдельности:
* В столбце `is_apartment` пропущены значения, скорее всего потому что данные за предыдущие года и на тот момент много кто не знал, что такое апартаменты и просто пропускали эту форму при заполнении обьявления. Исходя из этой логики, стоит заменить пропуски на значение "False", то есть "не апартаменты";
* В столбцах `parks_nearest` и `ponds_nearest` могут быть пропущены значения из-за того, что эти данные заполняются системой автоматически, в зависимости от местонахождения объекта недвижимости. Возможно, по техническим причинам эти данные система не заполнила. Т.к. данные столбцы не важны для анализа, пропуски трогать не будем, да и нет подходящего значения на замену;
* Столбец `balcony`. Логично предположить, что если не указано количество балконов, значит, скорее всего, их нет. Заполним пропуски на "0";
* Столбец `ceiling_height`. Пропуски могут быть связаны с незнанием высоты потолков владельцами недвижимости. Заполнять медианными значениями некорректно, потому что высота потолков в старом фонде жилья больше;
* Столбцы `airports_nearest`, `city_centers_nearest`, `parks_around3000` и `ponds_around3000`. Эти данные также заполняла система. Они могут быть пропущены из-за ошибки или из-за того, что нет подходящих объектов в нужном радиусе. Заполнять пропуски не будем;
* Столбец `days_exposition`. Пропуски возможно связаны с тем, что обьявление не было снято с публикации после продажи. Подходящих значений для замены пропусков нет;
* Столбцы `kitchen_area` и `living_area`. Эти значения могут быть не заполены владельцами недвижимости по причине незнания или нежелания. Эти столбцы также не трогаем;
* Столбец `floors_total` и `locality_name`.  Возможно, какой-то маленький процент людей просто пропустил или не увидел эти поля или не захотел заполнять. Таких значений пропущено немного, подходящих значений для замены нет.

Итак, заполним пропущенные значения в столбцах `is_apartment` и `balcony`.

#### Обработка пропусков в is_apartment

```python
# заменим значения на False
data['is_apartment'] = data['is_apartment'].fillna(False) 
```

```python
# проверим количество пропусков в столбцe
data['is_apartment'].isna().sum() 
```

#### Обработка пропусков в balcony

```python
# заменим значения на 0
data['balcony'] = data['balcony'].fillna(0) 
```

```python
# проверим количество пропусков в столбцe
data['balcony'].isna().sum() 
```

### Приведение данных в нужный тип

```python
# еще раз посмотрим на типы данных в столбцах
data.info()
```

```python
data.head(10)
```

Также рассмотрим каждый столбец по-отдельности:
* `last_price` можно привести к целочистенному типу, копейки не сыграют существенной роли при анализе, а также для удобства анализа можно разделить занчения на 1000;
* `first_day_exposition` нужно перевести в формат datetime, т.к. это дата;
* `is_apartment` теперь стал булевым, менять не нужно;
* `balcony` приведение к целочисленному типу;

```python
# разделим столбец со стоимостью на 1000 и округлим до целых
data['last_price'] = (data['last_price'] / 1000).round() 
```

```python
data['last_price'].head()
```

```python
# приведем данные в целочисленный тип
data[['last_price', 'balcony']] = data[['last_price', 'balcony']].astype('int') 
```

```python
# приведем данные в формат даты
data['first_day_exposition'] = pd.to_datetime(data['first_day_exposition'], format='%Y-%m-%d') 
```

```python
data.info()
```

### Вывод
Как видим, после предобработки осталось еще много незаполненных значений. Мы оставили их незаполненными по причине того, что заполнение их типовыми значениями(средним арифметическим или медианой) или константой(напрмер, нулями) может привести к некорректным реультатам анализа. Эти данные зависят от многих факторов, например, от удаленности от центра города, года постройки, т.к. старый фонд в г. Санкт-Петербурге существенно отличается по параметрам от типовой застройки. Поэтому, чтобы корректно заполнить пропуски, нужно провести отдельное исследование. Мы поработали с тем, что есть, постаравшись при этом не исказить данные. 

## Расчёты и добавление результатов в таблицу

```python
# посчитаем цену за квадратный метр и добавим столбец в data
data['price_of_m2'] = data['last_price'] / data['total_area'] 
data['price_of_m2'].head()
```

```python
# видим, что получился столбец с вещественными числами.
# Округлим значения и приведем к целочисленному типу
data['price_of_m2'] = data['price_of_m2'].round().astype('int')
data['price_of_m2'].head()
```

```python
# создадим столбец с днем недели публикации
data['weekday_of_exposition'] = data['first_day_exposition'].dt.weekday 
```

```python
# создадим столбец с месяцем публикации
data['month_of_exposition'] = data['first_day_exposition'].dt.month 
```

```python
# создадим столбец с годом публикации
data['year_of_exposition'] = data['first_day_exposition'].dt.year 
```

```python
# создадим функцию для категоризации стобца с этажами
def floor_cat(row):
    floor = row['floor']
    floors_total = row['floors_total']
    if floor == 1:
        return 'первый'
    if floor == floors_total:
        return 'последний'
    return 'другой'
```

```python
# применим функцию к датафрейму
data['floor_cat'] = data.apply(floor_cat, axis=1) 
```

```python
# создадим столбец отношения жилой площади к общей
data['living_to_total_ratio'] = data['living_area'] / data['total_area'] 
```

```python
# создадим столбец отношения площади кухни к общей
data['kitchen_to_total_ratio'] = data['kitchen_area'] / data['total_area'] 
```

```python
# проверим, как выглядят добавленные столбцы
data.head()
```

### Вывод

Для удобства дальнейшего анализа были произведены расчеты и добавлены столбцы в таблицу.
